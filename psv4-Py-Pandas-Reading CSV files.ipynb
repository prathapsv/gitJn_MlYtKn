## StringIO and BytesIO functions-IO Library

```python
# import io as io
from io import StringIO,BytesIO
```

```python
data=('Col1,Col2,Col3\n'
       'X,y,Z\n'
       'A,b,C\n'
       '1,2,3\n')
```

```python
type(data)
```

```python
type(StringIO())
```

```python
dfFromString=pd.read_csv(StringIO(data), usecols=['Col1','Col3'])#usecols=lambda x:x.upper() in ['Col1','Col3'])
```

```python
dfFromString.head()
```

```python
data2=('Col1,Col2,Col3,Col4\n'
       'X,5.567,Z,11\n'
       'A,88.898,C,22\n'
       'B,44.334,3,33\n')
```

```python
dfFromString2=pd.read_csv(StringIO(data2),dtype=object)
```

```python
dfFromString2
```

```python
dfFromString2['Col1']
```

```python
dfFromString2=pd.read_csv(StringIO(data2),dtype={'Col1':str,'Col2':float,'Col3':str,'col4':'Int64'})
```

```python
dfFromString2.info()
```

```python
dfFromString2
```

```python
data3=('Col1,Col2,Col3,IndexCol4\n'
       'X,5.567,Z,11\n'
       'A,88.898,C,22\n'
       'B,44.334,3,33\n')
```

```python
dfFromString2=pd.read_csv(StringIO(data3),dtype={'Col1':str,'Col2':float,'Col3':str,'col4':'Int64'})
```

```python
dfFromString2
```

```python
# Sets 4th column as index column..instead of default index column generated by data frame..like in above case..
dfFromString2=pd.read_csv(StringIO(data3),index_col=3,dtype={'Col1':str,'Col2':float,'Col3':str,'col4':'Int64'})
#index_col=3, is used to apply this change.
```

```python
dfFromString2
```

```python
data4=('a,b,c\n'
       '4,apple,Z,\n'
       '8,orange,C,\n'
      )
```

```python
dfFromString4=pd.read_csv(StringIO(data4))
```

```python
dfFromString4
#IMP NOte: By default when data is in above format, where first col has int values,pandas treats it as index col by default,
# so to avoid set index_col=False.. Note: This occurs only when each row's last column value is ending with a comma, ex:
# '4,apple,Z,\n' => where as if it ends properly like '4,apple,Z\n' then it won't matter..Default index is loaded instead of 
# treating first column as index column..
```

```python
dfFromString4=pd.read_csv(StringIO(data4),index_col=False)
```

```python
dfFromString4
```

```python
dfFromString4=pd.read_csv(StringIO(data4),index_col=False,usecols=['b','c'])
```

```python
dfFromString4
```

### Escape Chars in Data Frames

```python
# Escape Chars in Data Frames => escapechar='\\' 
data5=('a,b,c\n'
       '4,55,Z\n'
       '8,"hello \\"Will Smith\\" this is a long string for testing escapechar param",D\n'
      )
dfFromString5=pd.read_csv(StringIO(data5),escapechar='\\')
```

```python
dfFromString5
```
