```python
## simple examples
x=np.arange(0,10)
y=np.arange(11,21)
```

## plotting using matplotlib

### Scatter function

```python
# plotting using matplotlib
plt.scatter(x,y,c='g') #A scatter plot of *y* vs *x* with varying marker size and/or color.
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.show()
plt.savefig('E:\\SL\\ai\\test files\\matplotLibGraph.png')
```

### Plot function


```python
y=x*x
plt.plot(x,y,'y*',linestyle='dashed',linewidth=2,markersize=12)
# check function documentation for more customization options..
plt.xlabel('No of Years')
plt.ylabel('% of Population Growth')
plt.title('Population Growth')
# But seaborn has more options compared to this, just using for learning purpose, style implmn wise seaborn is better..
```

### SubPlot function

```python
plt.subplot(2,2,1) #(2,2,1)=> 1 indicates the index of the first subplot , 2nd and 3rd and so on..
plt.plot(x,y,'b')
plt.subplot(2,2,2)
plt.plot(x,y,'y--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
```

### Sin Chart using matplotlib

```python
np.pi
```

```python
# Sin Chart using matplotlib
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sin wave form')
plt.plot(x,y)
plt.show()
```

```python
# sine and cosine curves creation
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
# subplot with 2 as height(rows) and 1 as width(cols)
plt.subplot(2,1,1)

#first plot
plt.plot(x,y_sin,'b--')
plt.title('sine wave')

# second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine wave')
```

### Bar plots

```python
# Bar plots
x=[2,4,10]
y=[23,433,130]

x2=[3,9,18]
y2=[32,34,310]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('pop metrics')
```

## histograms

```python
# histogram gives a chart where it groups the elements into a range and then gives the count of those range..
x=[2,4,10,3,55,65,88,44,61,51,85,8,5,12,]
y=[1,2,3,4,55,66,99]

plt.subplot(2,1,1)
plt.hist(x)

plt.subplot(2,1,2)
plt.hist(y)

```

### Box plots

```python
#data2=np.random.normal(0,std,100)
#print(data2)
# read percentile in statistics vidoes to understand more on the box plot diagrams..
data = [np.random.normal(0,std,100) for std in range(1,4)] # check the usage of std.. in a loop as well as a param..
#print(data)
print(type(data))
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
```

### pie chart

```python
# data to plot
labels=['Python','C++','Java','C#']
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0) # explodes first slice

plt.pie(sizes,explode=explode,labels=labels,colors=colors, autopct='%1.1f%%', shadow=True)
plt.show()

plt.axis('equal')
plt.show()
```
