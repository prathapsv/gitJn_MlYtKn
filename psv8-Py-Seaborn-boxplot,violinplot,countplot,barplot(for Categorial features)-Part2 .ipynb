# Categorial Plots (to handle categorial features)
 boxplot,
 violinplot,
 countplot,
 barplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python

sns.countplot('day',data=df)
```

```python
sns.barplot(x='day',y='tip',data=df)
sns.barplot(x='smoker',y='tip',data=df)
sns.barplot(x='sex',y='tip',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,palette='rainbow',orient='v')
```

```python
sns.boxplot(y="day",x="total_bill",data=df,hue='smoker')
```

```python
sns.violinplot(y="day",x="total_bill",data=df,palette='rainbow')
```
